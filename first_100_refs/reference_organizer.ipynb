```python
counter = 0
zots = []
comment = []
x_counter = 0
majid_count = 0
for i in range(len(my_col_1)):
    list_of_items_1 = []
    list_of_items_2 = []
    try:
        if "$" in my_col_1[i]:
            majid_count = majid_count + 1
            list_of_items_1 = my_col_1[i].split("$")
            list_of_items_2 = my_col_2[i].split("$")
            #print(my_col_1[i])
            #print("")
            if len(list_of_items_1) == len(list_of_items_2):
                if my_col_1[i].count("zotero") == len(list_of_items_1):
                    #print("Hallo")
                    # fix the zots:
                    to_be_zots = []
                    for item in list_of_items_1:
                        after_item = item.split("/items/")[1]
                        to_be_zots.append(after_item[:8])
                    my_string = "_".join(to_be_zots)
                    zots.append(my_string)
                    counter = counter + 1
                    comment.append("ALLES_GUT_MULTIPLE")
                else:
                    print("GGGGGGG")
            else: 
                if my_col_1[i].count("zotero") == len(list_of_items_1) and len(list_of_items_1) < len(list_of_items_2): # not good
                    to_be_zots = []
                    for item in list_of_items_1:
                        after_item = item.split("/items/")[1]
                        to_be_zots.append(after_item[:8])
                    my_string = "_".join(to_be_zots)
                    zots.append(my_string)
                    counter = counter + 1
                    comment.append("ZOTEROS_GUT_EXTRA_CARE_NEEDED")
                    #print("Haiyah: ", my_col_1[i])
                    #print("FFFFFFF: ", my_col_2[i])
                else: # TAKE CARE here
                    print("Haiyah: ", my_col_1[i])
                    print("FFFFFFF: ", my_col_2[i])
                    print("HHHHHHHHHh")
        else:
            majid_count = majid_count + 1
            if my_col_1[i].count("zotero") == 1:
                after_item = item.split("/items/")[1]
                zots.append(after_item[:8])
                comment.append("ALLES_GUT_SINGLE")
                counter = counter + 1
            else: # two zoteros but no $ signs separating things properly
                #print("Haiyah: ", my_col_1[i])
                #print("FFFFFFF: ", my_col_2[i])
                #print("")
                comment.append("NO_ZOTERO_LINK")
                x_counter = x_counter + 1
            list_of_items_1.append(my_col_1[i])
            list_of_items_2.append(my_col_2[i])
            
            #print(my_col_1[i])
    except:
        majid_count = majid_count + 1
        list_of_items_1.append("NO_REFS_AVAILABLE")
        list_of_items_2.append("NO_REFS_AVAILABLE")
        counter = counter + 1


    #print(item)
print("Fuyouh: ", counter)
```
